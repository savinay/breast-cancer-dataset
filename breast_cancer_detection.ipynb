## Performing PCA on the dataset. 

### PCA is a statistical technique for dimensionality reduction and find patterns while trying to retain the variation in the data set. It helps us better visualize the dataset.

```python
X = data.iloc[:, 1:-1].values
Y = data.iloc[:, -1].values
```

```python
print(X.shape)
print(Y.shape)
```

```python
#calculating the number of eigenvalues that capture 50% of the variance.

sum_eig = 0
for val in eig_vals:
    sum_eig += val
sum_k_eig = 0
count = 0
for val in eig_vals:
    count += 1
    sum_k_eig += val
    if float(sum_k_eig)/float(sum_eig) > 0.5:
        break
print("First k eigenvalues that capture 50% energy: ", count)
```

```python
#visualizing the dataset after reducing the number of dimensions to 2

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_r = pca.fit(X_std).transform(X_std)
target_names = ['benign', 'malignant']
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
plt.figure()
colors = ['navy', 'turquoise']
lw = 2

for color, i, target_name in zip(colors, [0, 1], target_names):
    plt.scatter(X_r[Y == i, 0], X_r[Y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Breast Cancer dataset')
```

## We can see that the dataset can be split linearly between benign and malignant classes.

## Predicting the class using SVM

```python
#splitting the dataset into training and testing set (80%/20% split)

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std, Y, test_size=0.20, random_state=42)
```

```python
#applying SVM on the transformed dataset with PCA components from 2 to 11. 

from sklearn.svm import SVC
components = range(2,11)

for component in components:
    pca = PCA(n_components=component)
    pca.fit(X_train)
    X_t_train = pca.transform(X_train)
    X_t_test = pca.transform(X_test)
    clf = SVC()
    clf.fit(X_t_train, y_train)
    print('score with components: ', component, '= ', clf.score(X_t_test, y_test))
```

## We observe the highest accuracy with 2 PCA components by applying SVM on the dataset. 

### There can be 2 possibilities:
### 1. With more components, the model overfits the  data
### 2. There can be noise when we add more components

## Calculating the cross-validation accuracy on the dataset

```python
from sklearn.model_selection import cross_val_score
clf = SVC()
scores = cross_val_score(clf, X_std, Y, cv=5)
print(scores)
print("Cross-validation Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
